```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
file_path = r"C:\Users\afost\OneDrive\Desktop\Master in Fintech\Python\motor_insurance_recovery.csv"
df = pd.read_csv(file_path)

# Select categorical variables
categorical_cols = [
    "car_damage_severity",
    "hospital_visit_required",
    "hospital_admission_required",
    "rehabilitation_completed",
    "liability_admission_status"
]

# Reshape data for faceted plotting
df_melted = df[categorical_cols].melt(var_name="Variable", value_name="Category")

# Create faceted bar plots (small squares)
g = sns.catplot(
    data=df_melted,
    x="Category",
    col="Variable",
    kind="count",
    col_wrap=3,          # Wrap into multiple rows
    sharey=False,        # Independent y-axis for each plot
    palette="Set2",
    height=3.5,          # Square size
    aspect=1.2           # Slightly wider plots
)

# Improve readability
g.set_titles("{col_name}")
g.set_axis_labels("Category", "Count")
for ax in g.axes.flatten():
    ax.tick_params(axis='x', labelrotation=45, labelsize=9)
    ax.set_title(ax.get_title(), fontsize=11)

plt.tight_layout()
plt.show()
```
